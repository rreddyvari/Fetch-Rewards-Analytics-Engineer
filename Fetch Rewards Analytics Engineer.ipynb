```python
# convert all the date fields into the standard datetime format
# Defining a function called convert with an argument d

def convert(d):
    try:
        return(datetime.utcfromtimestamp(int(d['$date'])/1000).strftime('%Y-%m-%d %H:%M:%S'))
    except TypeError:
        return(None)
```

```python
#Converting all the datatime fields into the standard format

receipts_data['_id'] = receipts_data['_id'].apply(lambda x: x['$oid'])
receipts_data['createDate'] = receipts_data['createDate'].apply(lambda d: convert(d))
receipts_data['dateScanned'] = receipts_data['dateScanned'].apply(lambda d: convert(d))
receipts_data['finishedDate'] = receipts_data['finishedDate'].apply(lambda d: convert(d))
receipts_data['modifyDate'] = receipts_data['modifyDate'].apply(lambda d: convert(d))
receipts_data['pointsAwardedDate'] = receipts_data['pointsAwardedDate'].apply(lambda d: convert(d))
receipts_data['purchaseDate'] = receipts_data['purchaseDate'].apply(lambda d: convert(d))
```
